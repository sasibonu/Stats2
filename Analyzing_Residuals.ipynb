**PART A:**

Prove that $t = \frac{\widehat\beta_1}{\widehat{s.e.}(\widehat\beta_1)}$ can be written in the form of a t-distributed random variable with $n-2$ degrees of freedom. Specifically, let $Z$ be defined as $Z = \frac{\widehat\beta_1}{\sqrt{\frac{\sigma^2}{\sum^n_{i=1}\left(x_i - \bar{x}\right)^2}}}$, and $W = \frac{(n-2)\widehat\sigma^2}{\sigma^2}$. Show that

\begin{align*}
\frac{Z}{\sqrt{W\big/(n-2)}} = \frac{\widehat\beta_1}{\widehat{s.e.}(\widehat\beta_1)} = t  ,
\end{align*}

which, by definition, means that $t \sim t(n-2)$.

**Solution:**

To prove that $ t = \frac{\widehat{\beta}_1}{\widehat{s.e.}(\widehat{\beta}_1)}$ follows a t-distribution with $n-2$ degrees of freedom, we need to show that $\displaystyle \frac{Z}{\sqrt{W / (n-2)}} = t$.

Expanding $ \frac{Z}{\sqrt{W / (n-2)}}$:

\begin{align*}
\frac{Z}{\sqrt{W / (n-2)}} &= \frac{\widehat{\beta}_1}{\sqrt{\frac{\sigma^2}{\sum_{i=1}^n (x_i - \bar{x})^2}}}\cdot\frac{1}{\sqrt{\frac{(n-2)\widehat{\sigma}^2}{\sigma^2}}} \\
&= \frac{\widehat{\beta}_1}{\sqrt{\frac{\sigma^2}{\sum_{i=1}^n (x_i - \bar{x})^2}}}\cdot\frac{1}{\sqrt{\frac{(n-2)\widehat{\sigma}^2}{\sigma^2}}}\cdot\frac{\sqrt{\sum_{i=1}^n (x_i - \bar{x})^2}}{\sqrt{\sum_{i=1}^n (x_i - \bar{x})^2}} \\
&= \frac{\widehat{\beta}_1}{\frac{\sqrt{\sigma^2}}{\sqrt{\sum_{i=1}^n (x_i - \bar{x})^2}}}\cdot\sqrt{\frac{\sigma^2}{(n-2)\widehat{\sigma}^2}}\cdot\frac{\sqrt{\sum_{i=1}^n (x_i - \bar{x})^2}}{\sqrt{\sum_{i=1}^n (x_i - \bar{x})^2}} \\
&= \frac{\widehat{\beta}_1}{\frac{\sigma}{\sqrt{\sum_{i=1}^n (x_i - \bar{x})^2}}}\cdot\sqrt{\frac{\sigma^2}{(n-2)\widehat{\sigma}^2}}\cdot\frac{\sqrt{\sum_{i=1}^n (x_i - \bar{x})^2}}{\sqrt{\sum_{i=1}^n (x_i - \bar{x})^2}} \\
&= \frac{\widehat{\beta}_1}{\frac{\sigma}{\sqrt{\sum_{i=1}^n (x_i - \bar{x})^2}}}\cdot\sqrt{\frac{\sigma^2}{(n-2)\widehat{\sigma}^2}} \\
&= \frac{\widehat{\beta}_1}{\frac{\sigma}{\sqrt{\sum_{i=1}^n (x_i - \bar{x})^2}}}\cdot\sqrt{\frac{\sigma^2}{(n-2)\widehat{\sigma}^2}} \\
&= \frac{\widehat{\beta}_1}{\frac{\sigma}{\sqrt{(n-2)\frac{\widehat{\sigma}^2}{(n-2)}}}} \\
&= \frac{\widehat{\beta}_1}{\frac{\sigma}{\sqrt{(n-2)\frac{\widehat{\sigma}^2}{\widehat{\sigma}^2}}}} \\
&= \frac{\widehat{\beta}_1}{\frac{\sigma}{\sqrt{(n-2)}}} \\
&= \frac{\widehat{\beta}_1}{\text{se}(\widehat{\beta}_1)}
\end{align*}

This shows that $\displaystyle \frac{Z}{\sqrt{W / (n-2)}} = t$. Therefore, $\displaystyle t \sim t(n-2)$.

In class, we defined the *hat* or *projection* matrix as 

$$H = X(X^TX)^{-1}X^T.$$

The goal of this question is to use the hat matrix to prove that the fitted values, $\widehat{  Y}$, and the residuals, $\widehat{\boldsymbol\varepsilon}$, are uncorrelated. We will do it in steps.

**PART C:** 
Show that $H$ is symmetric: $H = H^T$.

**Solution:**

Given that $ {H} =  {X}( {X}^T {X})^{-1} {X}^T$,  its transpose is $ {H}^T$

$ {H}^T = ( {X}( {X}^T {X})^{-1} {X}^T)^T$

Since $( {AB})^T =  {B}^T {A}^T$:

${H}^T = ( {X}^T)^T(( {X}^T {X})^{-1})^T {X}^T$

Now, $( {X}^T)^T =  {X}$ and $(( {X}^T {X})^{-1})^T = ( {X}^T {X})^{-1}$

Hence,

$ {H}^T =  {X}( {X}^T {X})^{-1} {X}^T $ which equals $ {H} $

So, the hat matrix $ {H}$ is symmetric.


**PART D:** Show that $H(I_n - H) = 0_n$, where $0_n$ is the zero matrix of size $n \times n$.

**Solution:**


Given that, $H =  {X}( {X}^T {X})^{-1} {X}^T$ and $I_n$ is the $n \times n$ identity matrix, $H(I_n - H)$ is


$$  H(I_n - H) =  {X}( {X}^T {X})^{-1} {X}^T(I_n -  {X}( {X}^T {X})^{-1} {X}^T) $$
$$  = {X}( {X}^T {X})^{-1} {X}^T -  {X}( {X}^T {X})^{-1} {X}^T {X}( {X}^T {X})^{-1} {X}^T $$

$$  = {X}( {X}^T {X})^{-1} {X}^T -  {X}( {X}^T {X})^{-1} {X}^T $$

Hence, $H(I_n - H) =  {0}_n$.

## Problem 2 (15 points)

Researchers at the National Institutes of Standards and Technology (NIST) collected [pipline data](https://rdrr.io/cran/faraway/man/pipeline.html) on ultrasonic measurements of the depth of defects in the Alaska pipeline in the field. The depths of the defects were then remeasured in the laboratory. The laboratory measurements are more accurate than the field measurements, but more time consuming and expensive. We want to develop a regression model for correcting the in field measurements. 

**PART A:** Fit a regression model where `Lab` is the response and `Field` is the predictor and save this model as `lmodPipeline`. Check for non-constant variance. Use the Pearson residuals. The Pearson residuals are a type of standardized residual. A plot of the Pearson residuals against the fitted values provides evidence of nonconstant variance. You can do this by specifying the "type" argument in your use of the resid function. Note whether or not you see non-constant variance in your plot.

```r
install.packages('faraway', repos='http://cran.us.r-project.org')
library(faraway) #this package contains the pipeline data. 

data(pipeline); 
```

```r
lmodPipeline <- lm(Lab~Field, pipeline)

resid_pearson <- resid(lmodPipeline, type = "pearson")

plot(fitted(lmodPipeline), resid_pearson,
     xlab = "Fitted values",
     ylab = "Pearson residuals",
     main = "Pearson Residuals vs Fitted Values")
```

**Explanation**

The plot above shows non constant variance as the residuals are not constantly distributed. The plot shows that around the zero line the points are closer together at the lower fitted values. But as the fitted values increase, the points start to slowly move away from the zero line indicating an increase in the variance.  

**PART B:** Sometimes transforming the response and predictor helps in stabilizing variance. Find a transformation on Lab and/or Field so that in the transformed scale the relationship is approximately linear with constant variance. Restrict your choice of transformation to square root or log. Save your transformed variables as `pipeline$LabTransform` and `pipeline$FieldTransform`. Then, regress the transformed` Lab` variable (repsonse) onto the transformed `Field` variable (predictor), and save this as `lmodTr`.

```r
pipeline$LabTransform <- sqrt(pipeline$Lab)
pipeline$FieldTransform <- sqrt(pipeline$Field)

lmodTr <- lm(LabTransform ~ FieldTransform, pipeline)

resid_pearson_tr <- resid(lmodTr, type = "pearson")
fitted_values_tr <- predict(lmodTr)

plot(fitted_values_tr, resid_pearson, xlab = "Fitted Values (Transformed)",
     ylab = "Pearson Residuals (Transformed)", 
     main = "Transformed Pearson Residuals vs Fitted Values (Transformed)")

```

**Explanation**

Here, non constance variance can still be observed but it is better than the plot above. Here the points are not more spread out as the fitted values increase. The variance is more consistent which could mean that the square root modifications made the fitted values variance steady.

**Linearity**: Since there is no straight line that can approximate the data points, it can be said that this is non-linear. Rather, a curve would be a better fit for the data.
